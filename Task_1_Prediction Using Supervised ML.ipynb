## Simple Linear Regression Using Python Scikit Learn 

```python
# Importing all libraries required
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
```

### Importing the data from remote link

```python
url = "http://bit.ly/w-data"
data_table = pd.read_csv(url)
print("Successfully imported data into console")

data_table.head(25) 
```

### Plotting the distruibution of scores

```python
data_table.plot(x='Hours', y='Scores', style='o')    
plt.title('Hours vs Percentage')    
plt.xlabel('The Hours Studied')    
plt.ylabel('The Percentage Score')    
plt.show()
```

### Dividing the data into attributes and labels

```python
X = data_table.iloc[:, :-1].values    
y = data_table.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split    
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)   
```

### Training the Algorithm

```python
from sklearn.linear_model import LinearRegression    
regressor = LinearRegression()    
regressor.fit(X_train, y_train)   
  
print("Training ... Completed !.") 
```

### implementation of the plotting test data using the previously trained test data:

```python
line = regressor.coef_*X+regressor.intercept_  
plt.scatter(X, y)  
plt.plot(X, line);  
plt.show()  
```

### Predicting the scores for the model

```python
print(X_test)   
y_prediction = regressor.predict(X_test) 
```

### Comparing the actual versus predicted model to understand our model fitting

```python
cmp = pd.DataFrame({'Actual': y_test, 'Predicted': y_prediction})    
cmp  
```

### Evaluating the model

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_prediction))
```

### Testing the model
Testing our model with sample testing hours, so in this case, we take 9.5 hours, i.e, if a student studies for 9.25 hours, approximately how many marks he can get based on the data we received and the model we applied.

```python
hours = [[9.25]]  
our_prediction = regressor.predict(hours)  
print("Number of hours = {}".format(hours))  
print("Prediction Score = {}".format(our_prediction[0]))  
```
